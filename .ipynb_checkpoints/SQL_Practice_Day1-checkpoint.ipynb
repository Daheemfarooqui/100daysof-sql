# SQL Practice With DataFrames

Each question defines `df` (and `df1` where needed). Write your SQL in the next cell.

```python
from pandasql import sqldf
```

## Question 1
Find customers whose **total spending is greater than the average customer spending**.

```python
import pandas as pd
df=pd.DataFrame({'order_id':[1,2,3,4,5,6],'customer_id':[101,101,102,103,102,101],'amount':[500,700,300,1000,200,800]})
df
```

```python
sqldf("""
select customer_id , sum(amount) as total_spending
      from df 
      group by customer_id 
      having sum(amount) > (select avg(amount)
      from df 
      )

""")
```

## Question 2
Find the **2nd highest salary employee in each department**.

```python
import pandas as pd
df=pd.DataFrame({'emp_id':[1,2,3,4,5,6],'name':['A','B','C','D','E','F'],'department':['HR','HR','HR','IT','IT','IT'],'salary':[50000,60000,70000,40000,45000,80000]})
df
```

```python
sqldf("""
      
      select name , department , salary
        from  (select name , department , salary,
        dense_rank()
        over(partition by department order by salary desc)as rn 
        from df)
where rn = 2
      

""")
```

## Question 3
Find customers who **never placed an order**.

```python
import pandas as pd
df=pd.DataFrame({'customer_id':[101,102,103,104],'customer_name':['Alice','Bob','Charlie','David']})
df1=pd.DataFrame({'order_id':[1,2,3],'customer_id':[101,101,102]})
df,df1
```

```python
sqldf("""
select e.customer_name  
      from df e
      left join df1 o
      on e.customer_id = o.customer_id
      where o.order_id is null



""")
```

## Question 4
Find employees who **earn more than their manager**.

```python
import pandas as pd
df=pd.DataFrame({'emp_id':[1,2,3,4,5],'name':['A','B','C','D','E'],'manager_id':[None,1,1,2,2],'salary':[100000,60000,120000,70000,50000]})
df
```

```python
sqldf("""
      select e.name  as emp_name, e.salary , m.name as manager_name , m.salary
        from df e
        join df m on e.manager_Id = m.emp_id
        where e.salary > m.salary
      
      """)
```

## Question 5
Find products whose **price is greater than the average price of their category**.

```python
import pandas as pd
df=pd.DataFrame({'product_id':[1,2,3,4,5,6],'category':['A','A','A','B','B','B'],'price':[100,150,200,250,300,350]})
df
```

```python
sqldf("""
select product_id 
      from df a
      where price > (select avg(b.price)
                        from df b
                    where a.category = b.category)




""")
```

## Question 6
Find the **customer(s) with the maximum total spending**.

```python
import pandas as pd
df=pd.DataFrame({'order_id':[1,2,3,4,5,6],'customer_id':[101,101,102,103,102,101],'amount':[500,700,300,1000,200,800]})
df
```

```python
sqldf("""
        select customer_id , sum(amount) as max_spending
      from df
      group by customer_id
      having sum(amount) =  (select max(amount)
                              from (select sum(amount)as amount
                              from df 
                                  group by customer_id) 
      )



""")
```

## Question 7
Find employees whose **salary is greater than department average and less than department maximum**.

```python
import pandas as pd
df=pd.DataFrame({'emp_id':[1,2,3,4,5,6],'name':['A','B','C','D','E','F'],'department':['HR','HR','HR','IT','IT','IT'],'salary':[50000,60000,70000,40000,45000,80000]})
df
```

```python
sqldf("""
select name , salary 
      from df a
      where salary >(select avg(salary)
                        from df b
                        where a.department = b.department)
      and salary < (select max(salary) 
                    from df c
                        where a.department = c.department)







""")
```

## Question 8
For each product, return the **previous day's sales amount**.

```python
import pandas as pd
df=pd.DataFrame({'sale_id':[1,2,3,4,5],'product_id':['A','A','A','B','B'],'sale_date':['2025-01-01','2025-01-02','2025-01-03','2025-01-01','2025-01-02'],'amount':[100,150,120,200,250]})
df
```

```python
sqldf("""
select product_id , sale_date , amount
      from (select product_id ,sale_date , amount,
            dense_rank()
            over(partition by product_id order by sale_date desc)as rn
      from df)
where rn = 2


""")
```

## Question 9
For each customer, return their **most recent order**.

```python
import pandas as pd
df=pd.DataFrame({'order_id':[1,2,3,4,5,6],'customer_id':[101,101,102,103,102,101],'order_date':['2025-01-01','2025-01-15','2025-01-05','2025-01-10','2025-01-20','2025-02-01'],'amount':[500,700,300,1000,200,800]})
df
```

```python
sqldf("""
      select customer_id ,order_date 
        from (select customer_id , order_date ,
            dense_rank()
            over(partition by customer_id order by order_date desc)as rn
                from df )
      where rn = 1
      """)
```

## Question 10
Find the **2nd highest priced product in each category**.

```python
import pandas as pd
df=pd.DataFrame({'product_id':[1,2,3,4,5,6,7,8],'category':['A','A','A','B','B','B','C','C'],'price':[100,150,200,250,300,350,500,450]})
df
```

```python
sqldf("""
        select product_id , category ,price
        from (select product_id , category , price,
                dense_rank()
                    over(partition by category order by price desc)as rn
                from df ) t
        where rn = 2
      




""")
```
